# Pixel Processing

```python
from PIL import Image
image = Image.open('assets/white-rabbit-grass-1747425.jpg')
```

```python
from numpy import array
data1 = array(image)
data = data1
Image.fromarray(data)
```

## Darken the image

```python
def darken(data):
    first_dimension = len(data)
    second_dimension = len(data[0])
    third_dimesion = len(data[0][0])

    for i in range(first_dimension):
        for j in range(second_dimension):
            for k in range(third_dimesion):
                data[i][j][k]=max(data[i][j][k]-128,0)
    
    return data
```

```python
data_darken = darken(data)
Image.fromarray(data_darken)
```

## Brighten the image

```python
def brighten(data):
    first_dimension = len(data)
    second_dimension = len(data[0])
    third_dimesion = len(data[0][0])

    for i in range(first_dimension):
        for j in range(second_dimension):
            for k in range(third_dimesion):
                data[i][j][k]=min(data[i][j][k]+128,256)
    
    return data
```

```python
data_brighten = brighten(data)
Image.fromarray(data_brighten)
```

## Invert the colors

```python
def inverse(data):
    first_dimension = len(data)
    second_dimension = len(data[0])
    third_dimesion = len(data[0][0])

    for i in range(first_dimension):
        for j in range(second_dimension):
            for k in range(third_dimesion):
                data[i][j][k]=256-data[i][j][k]
    
    return data
```

```python
data_inverted = inverse(data)
Image.fromarray(data_inverted)
```
